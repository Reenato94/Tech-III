## 8) Consulta da quantidade de casos confirmados de COVID de acordo com a unidade federativa. Meses selecionados foram de Setembro a Novembro de 2020.

````python
#Seleção Geral
resultado_geral = spark.sql("SELECT unidade_federativa, count(resultado_teste) as qtd_positivos FROM tabela_temporaria WHERE resultado_teste = 'positivo' GROUP BY unidade_federativa, resultado_teste ORDER BY unidade_federativa")

resultado_geral_pd = resultado_geral.toPandas()

resultado_geral_pd = resultado_geral_pd.sort_values(by="qtd_positivos", ascending=False)

plt.figure(figsize=(10, 6))
axis=sns.barplot(x='qtd_positivos', y='unidade_federativa', data=resultado_geral_pd, palette='flare', hue='qtd_positivos')
axis.set_xlabel('Quantidade de pessoas que testaram positivo para COVID')
axis.set_ylabel('Unidade Federativa')
plt.title('Quantidade de pessoas que testaram positivo para COVID por Unidade Federativa - Setembro a Novembro 2020')

axis.xaxis.set_minor_locator(ticker.MultipleLocator(50))
axis.xaxis.set_major_locator(ticker.MultipleLocator(100))
plt.xlim(400,2300)
plt.grid(linestyle='--')

plt.show()
````

````python
# Criação de lista com unidades federativas top 5 em casos com resultado positivo em teste Covid
uf_top_5 = resultado_geral_pd.head(5)['unidade_federativa'].tolist()

list_top_5 = ','.join([f"'{val}'" for val in uf_top_5])  # Transforma a lista em string com valores formatados
````

## 9) Quantidade de casos separados se houve ou não sintomas

````python
#Filtro Sintomas
query = f"SELECT unidade_federativa, count(resultado_teste) as qtd_positivos, teve_sintomas \
          FROM tabela_temporaria \
          WHERE resultado_teste = 'positivo' AND unidade_federativa IN({list_top_5}) \
          GROUP BY  unidade_federativa, teve_sintomas"
resultado_sintomas = spark.sql(query)

sintomas_df = resultado_sintomas.toPandas()
sintomas_df = sintomas_df.sort_values(by="unidade_federativa")

plt.figure(figsize=(12, 6))  # Tamanho do gráfico
axis=sns.barplot(x='unidade_federativa', y='qtd_positivos', hue='teve_sintomas', data=sintomas_df, palette= "pastel")

# Títulos e rótulos
plt.title('Quantidade de Positivos por Unidade Federativa distribuídos por sintomas')
axis.set_xlabel('Unidade Federativa')
axis.set_ylabel('Quantidade de Positivos')

axis.yaxis.set_major_locator(ticker.MultipleLocator(100))
plt.ylim(0,2100)
plt.grid(linestyle='--')

plt.xticks(rotation=45)  # Rotacionar os nomes das unidades federativas para melhor visualização
plt.legend(title='Sintomas')
plt.tight_layout()

plt.show()
````

##10) Quantidade de pessoas que buscaram atendimento médico

````python
#Filtro se buscou atendimento médico
query = f"SELECT unidade_federativa, count(resultado_teste) as qtd_positivos, buscou_atendimento_medico \
          FROM tabela_temporaria \
          WHERE resultado_teste = 'positivo' AND unidade_federativa IN({list_top_5}) \
          GROUP BY  unidade_federativa, buscou_atendimento_medico"
resultado_atendimento = spark.sql(query)

atendimento_df = resultado_atendimento.toPandas()
atendimento_df = atendimento_df.sort_values(by="unidade_federativa")

plt.figure(figsize=(12, 6))  # Tamanho do gráfico
axis=sns.barplot(x='unidade_federativa', y='qtd_positivos', hue='buscou_atendimento_medico', data=atendimento_df, palette="pastel")

# Títulos e rótulos
plt.title('Quantidade de Positivos por Unidade Federativa se buscou atendimento médico ou não')
axis.set_xlabel('Unidade Federativa')
axis.set_ylabel('Quantidade de Positivos')

axis.yaxis.set_major_locator(ticker.MultipleLocator(100))
axis.yaxis.set_minor_locator(ticker.MultipleLocator(50))
#plt.ylim(0,2100)
plt.grid(linestyle='--')

plt.xticks(rotation=45)  # Rotacionar os nomes das unidades federativas para melhor visualização
plt.legend(title='Buscou atendimento médico')
plt.tight_layout()

plt.show()
````

  - Com a mesma população foram analisados quantos buscaram algum estabelecimento de saúde e/ou atendimento médico. A grande maioria que testou positivo não buscou nenhum das opções.
  - Considerando a população com casos em que testaram positivo para COVID e que buscaram algum tipo de atendimento médico, foi analisada a relação de casos em que o atendimento foi feito pelo SUS e os feitos de forma privada. Em todos os 5 Estados a maioria dos atendimentos ocorreram no SUS. Sendo que no Rio de Janeiro mesmo sendo a maioria pelo SUS, a quantidade de atendimentos privados foi quase semelhante.

Sobre comportamento da população em relação a restrição de contato os comportamentos foram semelhantes entre as unidades federativas analisadas:
  - A maioria das pessoas reduziu o contato com as pessoas, mas continuou saindo de casa para o trabalho ou atividades não essenciais e/ou recebendo visitas
  - Seguido das pessoas que ficaram em casa e só saiu em caso de necessidade básica
  - A minoria dos casos alegou ter ficado rigorosamente em casa, seguida da quantidade de pessoas que não aplicaram nenhum tipo de restrição


Quanto ao acesso a itens básico de limpeza e higiene, em todos os 5 Estados analisados a grande maioria possuía acesso.
  
Com base nos dados analisados e gráficos gerados, consideraria realizar as seguintes principais ações em um hospital num caso de um novo surto de COVID:
  - Independente da pessoa manifestar sintomas ou possuir comorbidades, se faz necessário realizar testes de covid em toda a população.
  - Como a maioria dos casos procurou atendimento no SUS, priorizaria a compra de testes com custo menor para garantir que todos que procurem atendimento consigam ser testados.
  - Ao identificar que algum paciente testou positivo, se faz necessário orientar os pacientes a evitarem contato o máximo possível, para evitar que novas pessoas se infectem.
Também focaria o atendimento do hospital para casos de população carente que recebe até 2 salários mínimos. Ofertando valores de tratamento mais acessíveis e com condições diferenciadas de pagamento.
